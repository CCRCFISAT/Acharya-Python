```python
import numpy as np
import math

x1 = np.arange(0,10)
y1 = x1**3

# y = x^3
plt.plot(x1, y1)

# y = x^2
plt.plot(x1, x1**2, "gx-")

# y = x^3 + x^2 * x
plt.plot(x1, x1**3 + x1**2 * x1, "co-")

# symbols  - , --, o , ^ , v , < , > , s , + , x , D , d , 1 , 2 , 3 , 4 , h , H , p , | , _
# colors   b, g, r, c, m, y, k, w
```

```python
x2 = np.arange(0, math.pi*2, 0.05)
plt.plot(x2, np.sin(x2), "r-")
plt.plot(x2, np.cos(x2), "m--")
```

# Bar chart
###### used for comparing discrete features or items

```python
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,39,12]

plt.bar(langs, students, color = "green")   
#barh() for horizontal

plt.ylabel("No. Students")
plt.show()
```

```python
divisions = ["A", "B", "C", "D", "E"]

# avg mark of a class
div_avg = np.random.randint(10, 50, len(divisions))

# avg mark of boys in class
boys_div_avg = np.random.randint(10, 50, len(divisions))

index = np.arange(len(divisions))
width = 0.3

plt.bar(index, div_avg, width, label="div avg")
plt.bar(index+width, boys_div_avg, width, color="red", label="boys avg")

plt.ylabel("Mark")
plt.xlabel("Division")
plt.legend()

plt.show()
```

# Scatter plot
###### used to represent relation between attributes. eg(below) cost vs performance

```python
# Create data
N = 30
cat1 = (np.random.uniform(0.1, 0.3, N)*10, np.random.randint(10, 25, N))
cat2 = (np.random.uniform(0.4, 0.6, N)*10, np.random.randint(25, 50, N))
cat3 = (np.random.uniform(0.7, 1.0, N)*10, np.random.randint(50, 100, N))

data = (cat1, cat2, cat3)
colors = ("red", "blue", "green")
groups = ("cat1", "cat2", "cat3")

# Create plot
# fig = plt.figure()
# ax = fig.add_subplot()

for data, color, group in zip(data, colors, groups):
    perf, cost = data
    plt.scatter(cost, perf, c=color,label=group)

plt.title('Performance vs Cost')
plt.xlabel("Cost")
plt.ylabel("Performance Index")
plt.legend()
plt.show()
```

# Pie Chart
###### useful for representing a single series of data

```python
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,39,12]
plt.pie(students, labels = langs)
plt.legend()
plt.show()
```

# subplots

```python
x = np.arange(1,50)
x_r = np.random.randn(100)

plt.subplots_adjust(wspace=0.5, hspace=0.5)

plt.subplot(2,2,1)
plt.title('histogram')
plt.hist(x_r)

plt.subplot(2,2,2)
plt.title('square root')
plt.plot(x,np.sqrt(x))

plt.subplot(2,2,3)
plt.title('exp')
plt.plot(x,np.exp(x))

plt.subplot(2,2,4)
plt.title('log')
plt.plot(x,np.log10(x))

plt.suptitle("subplots123")
plt.show()
```

###### Easier way to add subplots when there are many plots to be added

```python
fig, ax = plt.subplots(2,2)
ax = ax.reshape(-1)
plt.subplots_adjust(wspace=0.5, hspace=0.5)


# less no. of lines for each subplot 
ax[0].set_title('histogram')
ax[0].hist(x_r)

ax[1].set_title('exp')
ax[1].plot(x,np.exp(x))

ax[2].set_title('log')
ax[2].plot(x,np.log10(x))

ax[3].set_title('square root')
ax[3].plot(x,np.sqrt(x))

plt.suptitle("subplots124233")
plt.show()
```
